```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class UnorderedList:
    def __init__(self):
        self.head = None
    
    def isEmpty(self):
        return self.head == None
    
    def add(self, item):
        temp = Node(item)
        temp.next = self.head
        self.head = temp
    
    def size(self):
        current = self.head
        count = 0
        while current != None:
            count += 1
            current = current.next
        return count
    
    def search(self, item):
        current = self.head
        while current != None:
            if current.data == item:
                return True
            current = current.next
        return False
    
    def remove(self, item):
        current = self.head
        previous = None
        while current != None:
            if current.data == item:
                if previous == None:
                    self.head = current.next
                else:
                    previous.next = current.next
                return
            previous = current
            current = current.next
    
    def append(self, item):
        new = Node(item)
        if self.head == None:
            self.head = new
        else:
            current = self.head
            while current.next != None:
                current = current.next
            current.next = new
    
    def index(self, item):
        current = self.head
        pos = 0
        while current != None:
            if current.data == item:
                return pos
            current = current.next
            pos += 1
        return -1
    
    def insert(self, pos, item):
        new = Node(item)
        if pos == 0:
            new.next = self.head
            self.head = new
        else:
            current = self.head
            for i in range(pos - 1):
                current = current.next
            new.next = current.next
            current.next = new
    
    def pop(self, pos=None):
        if pos == None:
            if self.head == None:
                return None
            if self.head.next == None:
                data = self.head.data
                self.head = None
                return data
            current = self.head
            while current.next.next != None:
                current = current.next
            data = current.next.data
            current.next = None
            return data
        else:
            if pos == 0:
                data = self.head.data
                self.head = self.head.next
                return data
            current = self.head
            for i in range(pos - 1):
                current = current.next
            data = current.next.data
            current.next = current.next.next
            return data
    
    def __str__(self):
        if self.head == None:
            return "[]"
        result = []
        current = self.head
        while current != None:
            result.append(str(current.data))
            current = current.next
        return "[" + ", ".join(result) + "]"
    
    def slice(self, start, stop):
        new_list = UnorderedList()
        current = self.head
        pos = 0
        while current != None and pos < start:
            current = current.next
            pos += 1
        while current != None and pos < stop:
            new_list.append(current.data)
            current = current.next
            pos += 1
        return new_list
```

### Задание 3

Напишите функцию, которая принимает в качестве аргумента первый узел односвязного
списка, обращает порядок элементов списка (с разрушением исходного) и возвращает в
качестве результата первый узел полученного списка. Разработайте два варианта решения
этой задачи: 1) итеративный и 2) рекурсивный

```python
# Итеративный
def reverse_iterative(head):
    prev = None
    cur = head
    while cur:
        nxt = cur.next
        cur.next = prev
        prev = cur
        cur = nxt
    return prev

# Рекурсивный
def reverse_recursive(head, prev=None):
    if head is None:
        return prev
    nxt = head.next
    head.next = prev
    return reverse_recursive(nxt, head)

lst1 = UnorderedList()
for x in [1,2,3,4,5]:
    lst1.append(x)
print("Исходный:", lst1)
lst1.head = reverse_iterative(lst1.head)
print("После итеративного разворота:", lst1)

lst2 = UnorderedList()
for x in [1,2,3,4,5]:
    lst2.append(x)
print("\nИсходный:", lst2)
lst2.head = reverse_recursive(lst2.head)
print("После рекурсивного разворота:", lst2)
```

связанного списка. Например, для сравнения можно взять какую-либо операцию
выбранной линейной структуры

```python
# СТЕК (LIFO) — последний пришёл, первый ушёл
class Stack:
    def __init__(self):
        self.data = UnorderedList()
    
    def push(self, item):
        self.data.add(item)              # кладём в начало
    
    def pop(self):
        return self.data.pop(0)          # берём с конца
    
    def isEmpty(self):
        return self.data.isEmpty()       # проверка на пустоту
    
    def __str__(self):
        return str(self.data)


# ОЧЕРЕДЬ (FIFO) — первый пришёл, первый ушёл
class Queue:
    def __init__(self):
        self.data = UnorderedList()
    
    def enqueue(self, item):
        self.data.append(item)           # встаём в конец очереди 
    
    def dequeue(self):
        return self.data.pop(0)          # первый уходит 
    
    def isEmpty(self):
        return self.data.isEmpty()
    
    def __str__(self):
        return str(self.data)


# ДЕК — можно добавлять и удалять с обоих концов
class Deque:
    def __init__(self):
        self.data = UnorderedList()
    
    def addFront(self, item):
        self.data.add(item)              # добавляем в начало 
    
    def addRear(self, item):
        self.data.append(item)           # добавляем в конец
    
    def removeFront(self):
        return self.data.pop(0)          # удаляем из начала
    
    def removeRear(self):
        return self.data.pop()           # удаляем из конца
    
    def isEmpty(self):
        return self.data.isEmpty()
    
    def __str__(self):
        return str(self.data)

print("СТЕК")
s = Stack()
s.push(1); s.push(2); s.push(3)
print("push 1,2,3:", s)
print("pop:", s.pop())
print("после pop:", s)

print("\nОЧЕРЕДЬ:")
q = Queue()
q.enqueue("даша"); q.enqueue("игорь"); q.enqueue("пахан")
print("enqueue:", q)
print("dequeue:", q.dequeue())
print("после dequeue:", q)

print("\nДЕК:")
d = Deque()
d.addRear(10); d.addRear(20); d.addFront(5)
print("addRear(10,20) + addFront(5):", d)
print("removeFront:", d.removeFront())
print("removeRear:", d.removeRear())
print("после:", d)
```

### Задание 5

Реализуйте класс DoubleList для построения двухсвязных списков, в которых каждый узел
содержит ссылку на узел, предшествующий ему в списке, и ссылку на следующий узел.
Класс должен иметь следующие методы: проверка списка на пустоту, поиск элемента,
размер списка, вставка в начало, вставка в конец, удаление из начала, удаление из конца,
вставка перед указанным узлом, вставка после указанного узла, удаление указанного узла.
Проведите тестирование методов реализованного класса

```python
class DNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoubleList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def isEmpty(self):
        return self.head is None
    
    def size(self):
        cnt = 0
        cur = self.head
        while cur:                  # O(n) - проход по всем
            cnt += 1
            cur = cur.next
        return cnt
    
    def search(self, item):
        cur = self.head
        while cur:                  # O(n) - в худшем до конца
            if cur.data == item:
                return True
            cur = cur.next
        return False
    
    def addFront(self, item):
        new = DNode(item)
        if self.isEmpty():
            self.head = self.tail = new
        else:
            new.next = self.head    # новый указывает на старый head
            self.head.prev = new    # старый head указывает на новый
            self.head = new         # обновляем head
        # O(1) — просто меняем ссылки
    
    def addRear(self, item):
        new = DNode(item)
        if self.isEmpty():
            self.head = self.tail = new
        else:
            new.prev = self.tail    # новый указывает на старый хвост
            self.tail.next = new    # старый хвост указывает на новый
            self.tail = new         # обновляем хвост
    
    def removeFront(self):
        if self.isEmpty():
            return None
        data = self.head.data
        if self.head == self.tail:      # один элемент
            self.head = self.tail = None
        else:
            self.head = self.head.next  # двигаем голову
            self.head.prev = None       # у нового head prev = None
        return data                     # O(1)
    
    def removeRear(self):
        if self.isEmpty():
            return None
        data = self.tail.data
        if self.head == self.tail:      # один элемент
            self.head = self.tail = None
        else:
            self.tail = self.tail.prev  # двигаем хвост назад
            self.tail.next = None       # у нового хвоста next = None
        return data                     # O(1)
    
    def insertAfter(self, target, item):
        cur = self.head
        while cur and cur.data != target:  # O(n) — ищем target
            cur = cur.next
        if not cur:
            return
        new = DNode(item)
        if cur == self.tail:
            self.addRear(item)             #вставляем после хвоста
        else:
            new.prev = cur
            new.next = cur.next
            cur.next.prev = new
            cur.next = new
    
    def insertBefore(self, target, item):
        cur = self.head
        while cur and cur.data != target:  # O(n) — ищем target
            cur = cur.next
        if not cur:
            return
        if cur == self.head:
            self.addFront(item)            # если вставляем перед головой
        else:
            new = DNode(item)
            new.prev = cur.prev
            new.next = cur
            cur.prev.next = new
            cur.prev = new
    
    def removeNode(self, target):
        cur = self.head
        while cur and cur.data != target:  # O(n) — ищем target
            cur = cur.next
        if not cur:
            return None
        if cur == self.head:
            return self.removeFront()
        if cur == self.tail:
            return self.removeRear()
        cur.prev.next = cur.next           # перепрыгиваем
        cur.next.prev = cur.prev
        return cur.data
    
    def __str__(self):
        if self.isEmpty():
            return "[]"
        res = []
        cur = self.head
        while cur:                          # O(n) — обход всех
            res.append(str(cur.data))
            cur = cur.next
        return "[" + ", ".join(res) + "]"


dl = DoubleList()
dl.addFront(3); dl.addFront(2); dl.addFront(1)
dl.addRear(4); dl.addRear(5)
print("Список:", dl)

dl.insertAfter(3, 67)
print("insertAfter(3,67):", dl)

dl.insertBefore(4, 77)
print("insertBefore(4,77):", dl)

dl.removeNode(99)
print("removeNode(99):", dl)

print("removeFront:", dl.removeFront())
print("removeRear:", dl.removeRear())
print("search(3):", dl.search(3))
print("size:", dl.size())
```

### Задание 6

В HTML у каждого тега существуют открывающая и закрывающая формы, которые должны
быть сбалансированы, чтобы адекватно описывать веб-документ. Вот пример очень
простого веб-документа:
<html>
 <head>
 <title>
 Example
 </title>
 </head>
 <body>
 <h1>Hello, world</h1>
 </body>
</html>
Напишите программу, которая, используя структуру данных стек, проверит HTML
документ на соответствие открывающих и закрывающих тегов.
За основу решения можно взять алгоритм проверки скобочной последовательности из
лекционных материалов.

```python
def check_html_tags(html):
    """
    Проверяет HTML-документ на корректность тегов.
    Использует стек: открывающий тег кладём, закрывающий - проверяем.
    """
    stack = []
    i = 0
    errors = []
    
    while i < len(html):
        if html[i] == '<':
            j = i + 1
            while j < len(html) and html[j] != '>':
                j += 1
            
            if j < len(html):
                tag = html[i+1:j]
                
                # Пропускаем самозакрывающиеся теги типа <br/>
                if tag.endswith('/'):
                    i = j
                    continue
                
                if tag.startswith('/'):  # закрывающий тег
                    tag_name = tag[1:]
                    if not stack:
                        errors.append(f"Лишний закрывающий тег: </{tag_name}>")
                    elif stack[-1] != tag_name:
                        errors.append(f"Несоответствие: <{stack[-1]}> и </{tag_name}>")
                        stack.pop()
                    else:
                        stack.pop()
                else:  # открывающий тег
                    stack.append(tag)
            i = j
        i += 1
    
    # Проверяем незакрытые теги
    while stack:
        errors.append(f"Незакрытый тег: <{stack.pop()}>")
    
    return errors


filename = input("Введите путь к html файлу")

try:
    with open(filename, 'r', encoding='utf-8') as file:
        html_content = file.read()
    
    print(f"\n файл: {filename}")

    
    errors = check_html_tags(html_content)
    
    if errors:
        print("ошибки:")
        for err in errors:
            print(f"  - {err}")
    else:
        print("ВСЕ ТЕГИ КОРРЕКТНЫ")
        
except FileNotFoundError:
    print(f"Ошибка: файл '{filename}' не найден")
except Exception as e:
    print(f"Ошибка при чтении файла: {e}")
```

### Задание 7

Стековый калькулятор - это устройство (реальное или виртуальное), которое вычисляет
значения арифметических выражений, записанных в постфиксной форме (также она носит
название Обратная польская запись — Википедия (wikipedia.org)). Стековым он называется
так как действует по следующим правилам: если нам встречается число - мы должны
поместить его в стек; если знак операции - мы должны достать из стека столько чисел,
сколько необходимо для выполнения данной операции и выполнить ее.
Именно так и работает т.н. математический сопроцессор (например, в INtel8086+AMD64)-
часть ЦП, отвечающая за операции над числами с плавающей запятой.
Реализуйте стековый калькулятор. Напишите программу, которая читает выражение в
обратной польской нотации и выводит получившееся значение или пишет, что выражение
составлено не корректно (если оно некорректно).


```python
def calc_rpn(expr):
    stack = []
    for token in expr.split():
        if token in '+-*/':
            b = stack.pop()
            a = stack.pop()
            if token == '+': stack.append(a + b)
            elif token == '-': stack.append(a - b)
            elif token == '*': stack.append(a * b)
            elif token == '/':
                if b == 0: return "Ошибка: деление на ноль"
                stack.append(a / b)
        else:
            try:
                stack.append(float(token))
            except ValueError:
                return f"Ошибка: '{token}' не число"
    
    return stack[0] if len(stack) == 1 else f"Ошибка: в стеке {len(stack)} элементов"


expr = input("Введите выражение в обратной польской нотации: ")
result = calc_rpn(expr)
print(f"Результат: {result}")
```
