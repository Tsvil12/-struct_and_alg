### Задание 1
Возьмите реализацию класса UnorderedList из лекционных материалов и выполните следующие доработки:
1. Реализуйте оставшиеся методы этого класса append(item), index(item),
insert(pos,item), pop(), pop(pos). Требования к работе этих методов посмотрите в
лекционной презентации.
2. Реализуйте метод __str__, чтобы содержимое списка отображалось в Python-стиле (в
квадратных скобках).
3. Реализуйте метод “среза”. Он должен принимать два параметра (start и stop) и
возвращать копию списка, начиная с позиции start и заканчивая (но не включая!)
позицией stop.
Все выполненные доработки должны быть протестированны

```python
class Node:
    """Узел списка: хранит данные и ссылку на следующий узел"""
    def __init__(self, data):
        self.data = data
        self.next = None


class UnorderedList:
    def __init__(self):
        self.head = None  # ссылка на первый элемент (голову)
    
    def isEmpty(self):
        return self.head == None
    
    def add(self, item):
        """Добавляет в начало"""
        temp = Node(item)
        temp.next = self.head
        self.head = temp
    
    def size(self):
        """Возвращает количество элементов"""
        current = self.head
        count = 0
        while current != None:
            count += 1
            current = current.next
        return count
    
    def search(self, item):
        """Ищет элемент, возвращает True/False"""
        current = self.head
        while current != None:
            if current.data == item:
                return True
            current = current.next
        return False
    
    def remove(self, item):
        """Удаляет элемент"""
        current = self.head
        previous = None
        while current != None:
            if current.data == item:
                if previous == None:
                    self.head = current.next
                else:
                    previous.next = current.next
                return
            previous = current
            current = current.next
    
    # ========== НОВЫЕ МЕТОДЫ ==========
    
    def append(self, item):
        """Добавляет элемент в КОНЕЦ списка"""
        new = Node(item)
        if self.head == None:           # если список пуст
            self.head = new             # новый узел становится головой
        else:
            current = self.head
            while current.next != None: # идём до последнего узла
                current = current.next
            current.next = new          # последний узел ссылается на новый
    
    def index(self, item):
        """Возвращает ПОЗИЦИЮ элемента (считая с 0) или -1 если не найден"""
        current = self.head
        pos = 0
        while current != None:
            if current.data == item:
                return pos
            current = current.next
            pos += 1
        return -1  # не найден
    
    def insert(self, pos, item):
        """Вставляет элемент на позицию pos (0 = в начало)"""
        new = Node(item)
        
        if pos == 0:                     # вставка в начало
            new.next = self.head
            self.head = new
        else:
            current = self.head
            # доходим до узла, который будет ПЕРЕД новым
            for i in range(pos - 1):
                current = current.next
            new.next = current.next      # новый указывает на следующий
            current.next = new           # предыдущий указывает на новый
    
    def pop(self, pos=None):
        """
        Удаляет элемент и возвращает его.
        pop() - удаляет последний
        pop(pos) - удаляет элемент с позиции pos
        """
        if pos == None:      # удаляем ПОСЛЕДНИЙ элемент
            if self.head == None:
                return None
            if self.head.next == None:    # всего один элемент
                data = self.head.data
                self.head = None
                return data
            current = self.head
            # останавливаемся за ДВА узла до конца
            while current.next.next != None:
                current = current.next
            data = current.next.data
            current.next = None
            return data
        
        else:                 # удаляем элемент с позиции pos
            if pos == 0:      # удаляем первый элемент
                data = self.head.data
                self.head = self.head.next
                return data
            current = self.head
            # доходим до узла ПЕРЕД удаляемым
            for i in range(pos - 1):
                current = current.next
            data = current.next.data
            current.next = current.next.next  # перепрыгиваем удаляемый узел
            return data
    
    def __str__(self):
        """Выводит список в формате [1, 2, 3]"""
        if self.head == None:
            return "[]"
        result = []
        current = self.head
        while current != None:
            result.append(str(current.data))
            current = current.next
        return "[" + ", ".join(result) + "]"
    
    def slice(self, start, stop):
        """Возвращает НОВЫЙ список с элементами от start до stop (не включая stop)"""
        new_list = UnorderedList()
        current = self.head
        pos = 0
        
        # пропускаем элементы до start
        while current != None and pos < start:
            current = current.next
            pos += 1
        
        # добавляем элементы от start до stop-1
        while current != None and pos < stop:
            new_list.append(current.data)
            current = current.next
            pos += 1
        
        return new_list


print("Тест")

lst = UnorderedList()
lst.add(30)
lst.add(20)
lst.add(10)
print(f"После add(30), add(20), add(10): {lst}")

lst.append(40)
lst.append(50)
print(f"После append(40), append(50): {lst}")

print(f"index(30) = {lst.index(30)}")
print(f"index(100) = {lst.index(100)} (не найден)")

lst.insert(2, 99)
print(f"После insert(2, 99): {lst}")

print(f"pop() = {lst.pop()}")
print(f"После pop(): {lst}")

print(f"pop(2) = {lst.pop(2)}")
print(f"После pop(2): {lst}")

print(f"slice(1, 4): {lst.slice(1, 4)}")
```

связном списке без повторов. Пользователь последовательно вводит с клавиатуры n строк.
При вводе строки, которая раньше не встречалась, вставьте ее в начало списка. При вводе
уже внесенной строки удалите ее из списка и снова вставьте в начало.
Такая функция реализует стратегию «сдвиг в начало», которая применяется в кешировании,
при сжатии данных и т.д.


```python
def move_to_front():
    """
    Реализует стратегию "сдвиг в начало" (move-to-front).
    
    Как работает:
    - Если строка новая - вставляем в начало списка
    - Если строка уже была - удаляем её и снова вставляем в начало
    - Так часто используемые элементы всегда оказываются в начале
    
    Возвращает: итоговый связный список
    """
    lst = UnorderedList()           # создаём пустой список
    n = int(input("Введите количество строк: "))
    
    for i in range(n):
        s = input(f"Строка {i+1}: ")
        
        if lst.search(s):           # если строка уже есть в списке
            lst.remove(s)           # удаляем её
            print(f"'{s}' уже была, удаляем и вставляем в начало")
        else:
            print(f"'{s}' новая, вставляем в начало")
        
        lst.add(s)                  # вставляем в начало (и для новых, и для старых)
        print(f"  Текущий список: {lst}\n")
    
    print(f"\nИтоговый список: {lst}")
    return lst


# ========== ЗАПУСК ==========
print("тест")
move_to_front()
```

### Задание 3

Напишите функцию, которая принимает в качестве аргумента первый узел односвязного
списка, обращает порядок элементов списка (с разрушением исходного) и возвращает в
качестве результата первый узел полученного списка. Разработайте два варианта решения
этой задачи: 1) итеративный и 2) рекурсивный

```python
def reverse_list(head):
    """
    Разворачивает односвязный список (in-place, с разрушением исходного).
    
    Принцип работы:
    Проходим по списку и у каждого узла разворачиваем ссылку 
    в противоположную сторону.
    
    Было:  head → [1] → [2] → [3] → None
    Стало: head → [3] → [2] → [1] → None
    
    Параметры:
        head - первый узел исходного списка
    
    Возвращает:
        новый первый узел (старый последний)
    """
    prev = None          # предыдущий узел (станет новым next)
    current = head       # текущий обрабатываемый узел
    
    while current != None:
        nxt = current.next    # сохраняем ссылку на следующий (иначе потеряем список!)
        current.next = prev   # разворачиваем ссылку: теперь current указывает на предыдущий
        prev = current        # сдвигаем prev на текущий
        current = nxt         # переходим к следующему узлу
    
    return prev  # prev теперь указывает на новый первый узел


# Ввод данных от пользователя
lst = UnorderedList()
n = int(input("\nВведите количество элементов в списке: "))

print("Введите элементы списка (каждый с новой строки):")
for i in range(n):
    value = input(f"  Элемент {i+1}: ")
    lst.append(value)

print(f"\nИсходный список: {lst}")
print("  (связи: элементы идут в порядке ввода)")

# Разворачиваем
lst.head = reverse_list(lst.head)

print(f"\перевернутый список: {lst}")
print("  (связи: элементы идут в обратном порядке)")

# Проверка на пустом списке (если пользователь ввёл 0)
if n == 0:
    print("\n Был введён пустой список")
```

### Задание 4

Используя реализацию связного списка из задания № 1 (класс UnorderedList), реализуйте
следующие линейные структуры (необходимый набор методов посмотрите в лекционной
презентации):
1. Стек.
2. Очередь.
3. Дек.

Проведите тестирование реализованных линейных структур.
Продумайте и проведите показательный эксперимент, сравнивающий производительность
стека, очереди или дека, основанных на списках Python, с реализациями на основе
связанного списка. Например, для сравнения можно взять какую-либо операцию
выбранной линейной структуры

```python
class Stack:
    """
    Стек - структура LIFO (Last In - First Out).
    Последний добавленный элемент будет удалён первым.
    
    Пример: стопка тарелок, история браузера (кнопка "Назад")
    
    Операции (все O(1)):
    - push() - положить на вершину
    - pop()  - взять с вершины
    - peek() - посмотреть вершину без удаления
    """
    def __init__(self):
        self.data = UnorderedList()   # внутри используем связный список
    
    def isEmpty(self):
        return self.data.isEmpty()
    
    def push(self, item):
        """Добавляем элемент на вершину стека (в начало списка)"""
        self.data.add(item)           # add() - O(1)
    
    def pop(self):
        """Удаляем и возвращаем элемент с вершины стека (из начала)"""
        if self.isEmpty():
            return None
        return self.data.pop(0)       # pop(0) - удаление первого элемента
    
    def peek(self):
        """Смотрим верхний элемент, не удаляя его"""
        if self.isEmpty():
            return None
        return self.data.head.data    # просто читаем данные из головы
    
    def size(self):
        return self.data.size()
    
    def __str__(self):
        return str(self.data)

class Queue:
    """
    Очередь - структура FIFO (First In - First Out).
    Первый добавленный элемент будет удалён первым.
    
    Пример: очередь в магазине, печать документов на принтере
    
    Операции:
    - enqueue() - добавить в конец очереди (O(n))
    - dequeue() - удалить из начала очереди (O(1))
    """
    def __init__(self):
        self.data = UnorderedList()
    
    def isEmpty(self):
        return self.data.isEmpty()
    
    def enqueue(self, item):
        """Добавляем элемент в КОНЕЦ очереди"""
        self.data.append(item)        # append() - O(n)
    
    def dequeue(self):
        """Удаляем элемент из НАЧАЛА очереди"""
        if self.isEmpty():
            return None
        return self.data.pop(0)       # pop(0) - O(1)
    
    def size(self):
        return self.data.size()
    
    def __str__(self):
        return str(self.data)


class Deque:
    """
    Дек (Double-Ended Queue) - двусторонняя очередь.
    Можно добавлять и удалять элементы с ОБОИХ концов.
    
    Пример: очередь в банке, где можно зайти и с начала, и с конца
    """
    def __init__(self):
        self.data = UnorderedList()
    
    def isEmpty(self):
        return self.data.isEmpty()
    
    def addFront(self, item):
        """Добавляем элемент в НАЧАЛО дека"""
        self.data.add(item)           # add() - O(1)
    
    def addRear(self, item):
        """Добавляем элемент в КОНЕЦ дека"""
        self.data.append(item)        # append() - O(n)
    
    def removeFront(self):
        """Удаляем элемент из НАЧАЛА дека"""
        if self.isEmpty():
            return None
        return self.data.pop(0)       # pop(0) - O(1)
    
    def removeRear(self):
        """Удаляем элемент из КОНЦА дека"""
        if self.isEmpty():
            return None
        return self.data.pop()        # pop() - O(n) (поиск последнего)
    
    def size(self):
        return self.data.size()
    
    def __str__(self):
        return str(self.data)


print("Итог")

print("\n--- 1. СТЕК (LIFO) ---")
print("Принцип: последний пришёл - первый ушёл (как стопка тарелок)")

s = Stack()
print(f"Создан пустой стек: {s}")
print(f"isEmpty() = {s.isEmpty()}")

s.push(10)
s.push(20)
s.push(30)
print(f"\nПосле push(10), push(20), push(30): {s}")
print(f"size() = {s.size()}")

print(f"\npeek() = {s.peek()} (просто смотрим верхний элемент)")
print(f"pop() = {s.pop()} (забираем верхний)")
print(f"После pop(): {s}")
print(f"pop() = {s.pop()}")
print(f"pop() = {s.pop()}")
print(f"После трёх pop(): {s}")
print(f"isEmpty() = {s.isEmpty()}")

print("\n--- 2. ОЧЕРЕДЬ (FIFO) ---")
print("Принцип: первый пришёл - первый ушёл (как очередь в магазине)")

q = Queue()
q.enqueue("Анна")
q.enqueue("Борис")
q.enqueue("Виктор")
print(f"\nПосле enqueue('Анна'), enqueue('Борис'), enqueue('Виктор'): {q}")
print(f"size() = {q.size()}")

print(f"\ndequeue() = {q.dequeue()} (уходит первый в очереди)")
print(f"После dequeue(): {q}")
print(f"dequeue() = {q.dequeue()}")
print(f"dequeue() = {q.dequeue()}")
print(f"После всех dequeue(): {q}")


print("\n--- 3. ДЕК (двусторонняя очередь) ---")
print("Принцип: можно добавлять и удалять с обоих концов")

d = Deque()
print(f"Создан пустой дек: {d}")

d.addRear(10)
d.addRear(20)
print(f"\nПосле addRear(10), addRear(20): {d}")
d.addFront(5)
print(f"После addFront(5): {d}")
d.addFront(1)
print(f"После addFront(1): {d}")

print(f"\nremoveFront() = {d.removeFront()} (удаляем из начала)")
print(f"removeRear() = {d.removeRear()} (удаляем из конца)")
print(f"После удалений: {d}")
print(f"size() = {d.size()}")


print("СРАВНЕНИЕ ПРОИЗВОДИТЕЛЬНОСТИ: связный список vs список Python")

import time

def compare_performance():
    """
    Сравниваем скорость работы стека на связном списке 
    и на встроенном списке Python
    """
    n = 10000  # количество операций
    
    # ТЕСТ 1: Стек на связном списке
    stack_linked = Stack()
    start = time.time()
    for i in range(n):
        stack_linked.push(i)
    for i in range(n):
        stack_linked.pop()
    linked_time = time.time() - start
    
    # ТЕСТ 2: Стек на списке Python
    stack_py = []
    start = time.time()
    for i in range(n):
        stack_py.append(i)      # аналог push
    for i in range(n):
        stack_py.pop()          # аналог pop
    py_time = time.time() - start
    
    print(f"\nКоличество операций: {n} push + {n} pop = {n*2} операций")
    print(f"Стек на СВЯЗНОМ СПИСКЕ: {linked_time:.4f} секунд")
    print(f"Стек на СПИСКЕ PYTHON:  {py_time:.4f} секунд")
    
    if linked_time > py_time:
        print(f"\nСписок Python работает БЫСТРЕЕ в {linked_time/py_time:.1f} раз")
    else:
        print(f"\nСвязный список работает БЫСТРЕЕ в {py_time/linked_time:.1f} раз")
    
    

compare_performance()
```

### Задание 5

Реализуйте класс DoubleList для построения двухсвязных списков, в которых каждый узел
содержит ссылку на узел, предшествующий ему в списке, и ссылку на следующий узел.
Класс должен иметь следующие методы: проверка списка на пустоту, поиск элемента,
размер списка, вставка в начало, вставка в конец, удаление из начала, удаление из конца,
вставка перед указанным узлом, вставка после указанного узла, удаление указанного узла.
Проведите тестирование методов реализованного класса

```python
class DNode:
    """Узел двусвязного списка"""
    def __init__(self, data):
        self.data = data
        self.next = None   # ссылка на следующий
        self.prev = None   # ссылка на предыдущий


class DoubleList:
    def __init__(self):
        self.head = None   # голова
        self.tail = None   # хвост
    
    def isEmpty(self):
        return self.head == None
    
    def size(self):
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count
    
    def search(self, item):
        current = self.head
        while current:
            if current.data == item:
                return True
            current = current.next
        return False
    
    def addFront(self, item):
        new = DNode(item)
        if self.isEmpty():
            self.head = self.tail = new
        else:
            new.next = self.head
            self.head.prev = new
            self.head = new
    
    def addRear(self, item):
        new = DNode(item)
        if self.isEmpty():
            self.head = self.tail = new
        else:
            new.prev = self.tail
            self.tail.next = new
            self.tail = new
    
    def removeFront(self):
        if self.isEmpty():
            return None
        data = self.head.data
        if self.head == self.tail:
            self.head = self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = None
        return data
    
    def removeRear(self):
        if self.isEmpty():
            return None
        data = self.tail.data
        if self.head == self.tail:
            self.head = self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None
        return data
    
    def insertBefore(self, target, item):
        """Вставка перед элементом с данными target"""
        current = self.head
        while current and current.data != target:
            current = current.next
        if not current:
            return
        
        new = DNode(item)
        if current == self.head:
            self.addFront(item)
        else:
            new.prev = current.prev
            new.next = current
            current.prev.next = new
            current.prev = new
    
    def insertAfter(self, target, item):
        """Вставка после элемента с данными target"""
        current = self.head
        while current and current.data != target:
            current = current.next
        if not current:
            return
        
        new = DNode(item)
        if current == self.tail:
            self.addRear(item)
        else:
            new.prev = current
            new.next = current.next
            current.next.prev = new
            current.next = new
    
    def removeNode(self, target):
        """Удаление элемента с данными target"""
        current = self.head
        while current and current.data != target:
            current = current.next
        if not current:
            return None
        
        if current == self.head:
            return self.removeFront()
        if current == self.tail:
            return self.removeRear()
        
        current.prev.next = current.next
        current.next.prev = current.prev
        return current.data
    
    def __str__(self):
        if self.isEmpty():
            return "[]"
        res = []
        cur = self.head
        while cur:
            res.append(str(cur.data))
            cur = cur.next
        return "[" + ", ".join(res) + "]"


print("Двусвязный список")
dl = DoubleList()

dl.addFront(3)
dl.addFront(2)
dl.addFront(1)
dl.addRear(4)
dl.addRear(5)
print(f"Список: {dl}")

dl.insertAfter(3, 99)
print(f"После insertAfter(3, 99): {dl}")

dl.insertBefore(4, 77)
print(f"После insertBefore(4, 77): {dl}")

dl.removeNode(99)
print(f"После removeNode(99): {dl}")

print(f"removeFront(): {dl.removeFront()}")
print(f"После removeFront(): {dl}")

print(f"removeRear(): {dl.removeRear()}")
print(f"После removeRear(): {dl}")

print(f"search(3): {dl.search(3)}")
print(f"size(): {dl.size()}")
```

### Задание 6

В HTML у каждого тега существуют открывающая и закрывающая формы, которые должны
быть сбалансированы, чтобы адекватно описывать веб-документ. Вот пример очень
простого веб-документа:
<html>
 <head>
 <title>
 Example
 </title>
 </head>
 <body>
 <h1>Hello, world</h1>
 </body>
</html>
Напишите программу, которая, используя структуру данных стек, проверит HTML
документ на соответствие открывающих и закрывающих тегов.
За основу решения можно взять алгоритм проверки скобочной последовательности из
лекционных материалов.

```python
def check_html_tags(html):
    """
    Проверяет HTML-документ на корректность тегов.
    Использует стек: открывающий тег кладём, закрывающий - проверяем.
    """
    stack = []
    i = 0
    errors = []
    
    while i < len(html):
        if html[i] == '<':
            j = i + 1
            while j < len(html) and html[j] != '>':
                j += 1
            
            if j < len(html):
                tag = html[i+1:j]
                
                # Пропускаем самозакрывающиеся теги типа <br/>
                if tag.endswith('/'):
                    i = j
                    continue
                
                if tag.startswith('/'):  # закрывающий тег
                    tag_name = tag[1:]
                    if not stack:
                        errors.append(f"Лишний закрывающий тег: </{tag_name}>")
                    elif stack[-1] != tag_name:
                        errors.append(f"Несоответствие: <{stack[-1]}> и </{tag_name}>")
                        stack.pop()
                    else:
                        stack.pop()
                else:  # открывающий тег
                    stack.append(tag)
            i = j
        i += 1
    
    # Проверяем незакрытые теги
    while stack:
        errors.append(f"Незакрытый тег: <{stack.pop()}>")
    
    return errors




# Чтение HTML из файла
filename = input("Введите путь к html файлу")

try:
    with open(filename, 'r', encoding='utf-8') as file:
        html_content = file.read()
    
    print("\nФайл прочитан. Проверяю теги...")
    print("-" * 50)
    
    errors = check_html_tags(html_content)
    
    if errors:
        print("НАЙДЕНЫ ОШИБКИ:")
        for err in errors:
            print(f"  - {err}")
    else:
        print("ВСЕ ТЕГИ КОРРЕКТНЫ")
        
except FileNotFoundError:
    print(f"Ошибка: файл '{filename}' не найден")
except Exception as e:
    print(f"Ошибка при чтении файла: {e}")
```

### Задание 7

Стековый калькулятор - это устройство (реальное или виртуальное), которое вычисляет
значения арифметических выражений, записанных в постфиксной форме (также она носит
название Обратная польская запись — Википедия (wikipedia.org)). Стековым он называется
так как действует по следующим правилам: если нам встречается число - мы должны
поместить его в стек; если знак операции - мы должны достать из стека столько чисел,
сколько необходимо для выполнения данной операции и выполнить ее.
Именно так и работает т.н. математический сопроцессор (например, в INtel8086+AMD64)-
часть ЦП, отвечающая за операции над числами с плавающей запятой.
Реализуйте стековый калькулятор. Напишите программу, которая читает выражение в
обратной польской нотации и выводит получившееся значение или пишет, что выражение
составлено не корректно (если оно некорректно).


```python
def rpn_calculator():
    """
    Стековый калькулятор с пошаговым объяснением
    """
    stack = []
    
    print("СТЕКОВЫЙ КАЛЬКУЛЯТОР")
    print("\nПравила:")
    print("  - Числа просто кладутся в стек")
    print("  - Операции (+ - * /) забирают два числа из стека")
    print("  - Результат операции кладётся обратно в стек")
    
    expression = input("\nВведите выражение: ")
    tokens = expression.split()
    
    print("ВЫПОЛНЕНИЕ:") 
    
    for token in tokens:
        if token in ['+', '-', '*', '/']:
            # Это операция
            if len(stack) < 2:
                print(f"\nОШИБКА: Для операции '{token}' нужно 2 числа")
                return None
            
            b = stack.pop()
            a = stack.pop()
            
            if token == '+':
                res = a + b
                print(f"  {a} + {b} = {res}")
            elif token == '-':
                res = a - b
                print(f"  {a} - {b} = {res}")
            elif token == '*':
                res = a * b
                print(f"  {a} * {b} = {res}")
            elif token == '/':
                if b == 0:
                    print(f"\nОШИБКА: Деление на ноль")
                    return None
                res = a / b
                print(f"  {a} / {b} = {res}")
            
            stack.append(res)
            print(f"    Стек: {stack}")
        else:
            # Это число
            try:
                if '.' in token:
                    num = float(token)
                else:
                    num = int(token)
                stack.append(num)
                print(f"  Кладём {num} -> стек {stack}")
            except ValueError:
                print(f"\nОШИБКА: '{token}' - не число и не операция")
                return None
    
    
    if len(stack) == 0:
        print("ОШИБКА: Вы ничего не ввели")
        return None
    elif len(stack) > 1:
        print(f"ОШИБКА: В стеке осталось {len(stack)} чисел: {stack}")
        print("Значит, в выражении не хватает операций")
        return None
    else:
        print(f"ИТОГОВЫЙ РЕЗУЛЬТАТ: {stack[0]}")
        return stack[0]


rpn_calculator()
```

### Задание 8

Используя структуру данных очередь и пример из лекционных материалов, создайте
симуляцию ситуации из реальной жизни согласно своему варианту:

### Вариант 2. Покупатели в очереди на кассе бакалейного магазина

```python
import random

print("Симуляция очереди")

# Ввод данных
total_customers = int(input("Сколько покупателей пришло за 10 минут: "))

print("\nВыберите скорость кассира:")
print("  1 - МЕДЛЕННО (2-4 минуты на человека)")
print("  2 - НОРМАЛЬНО (1-2 минуты)")
print("  3 - БЫСТРО (0.5-1 минута)")
speed = int(input("Введите 1, 2 или 3: "))

# Задаём время обслуживания в секундах
if speed == 1:
    min_time = 120   # 2 минуты
    max_time = 240   # 4 минуты
    speed_name = "МЕДЛЕННО"
elif speed == 2:
    min_time = 60    # 1 минута
    max_time = 120   # 2 минуты
    speed_name = "НОРМАЛЬНО"
else:
    min_time = 30    # 30 секунд
    max_time = 60    # 1 минута
    speed_name = "БЫСТРО"

print(f"РЕЗУЛЬТАТ:")
print(f"  Пришло покупателей: {total_customers}")
print(f"  Скорость кассира: {speed_name}")

# Симуляция
queue = 0           # очередь
served = 0          # обслужено
cash_busy = 0       # касса занята (секунд)

# 10 минут = 600 секунд
for sec in range(600):
    # Покупатели приходят равномерно в течение 10 минут
    if sec < 600:
        # Распределяем приход покупателей равномерно
        if sec * total_customers // 600 > (sec - 1) * total_customers // 600:
            queue += 1
    
    # Касса свободна и есть очередь
    if cash_busy == 0 and queue > 0:
        queue -= 1
        served += 1
        cash_busy = random.randint(min_time, max_time)
    
    # Касса работает
    if cash_busy > 0:
        cash_busy -= 1

# Итоги
print(f"  Обслужено покупателей: {served}")
print(f"  Осталось в очереди: {queue}")

if queue == 0:
    print("\nОчереди нет, кассир справляется")
elif queue < total_customers // 2:
    print("\nНебольшая очередь, но терпимо")
else:
    print("\nБольшая очередь, люди недовольны")
```
